```python
class Celular:
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} se ha encendido.")

    def mostrar_info(self):
        print(f"Celular: {self.marca} {self.modelo}, Precio: ${self.precio}")

   # Creacion de objetos (instacias)
    celular1 = Celular("Samsung", "Galaxy S23", 800)
    celular2 = Celular("Apple", "iPhone 15",1000)
   # Llamadas a las funciones por cada objeto celular 1 y celylar 2
    print("Celular 1")
    celular1.mostrar_info()
    celular2.encender()

    print("/nCelular 2")
    celular1.mostrar_info()
    celular2.encender()


    print("Programa terminado")
    print("Creador de software por Hernadez Dylan Nl = 24")
```
